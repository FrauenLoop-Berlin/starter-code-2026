## A quick JS ↔ Python cheat sheet

You've written JavaScript in the Flask web app (assignment 2.4 — checkout button). Here's the mapping:

| Concept | JavaScript | Python |
|---|---|---|
| Variable | `let price = 3.5;` | `price = 3.5` |
| Array / List | `[1, 2, 3]` | `[1, 2, 3]` |
| Object / Dict | `{name: 'Latte', price: 3.5}` | `{'name': 'Latte', 'price': 3.5}` |
| If | `if (x > 10) { ... }` | `if x > 10:` |
| Loop | `for (let x of arr) { ... }` | `for x in arr:` |
| Block scope | `{ }` braces | indentation (4 spaces) |
| Statement end | `;` | newline |

**Python is dynamically typed** (like JS) but **strict about indentation** (unlike JS).
