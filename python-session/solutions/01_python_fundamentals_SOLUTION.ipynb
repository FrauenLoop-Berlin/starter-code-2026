# Exercise 1 — SOLUTION

Café Cozy order calculator with worked answers + short teaching notes.

```python
order = [
    {'name': 'Latte',            'price': 3.75, 'quantity': 2},
    {'name': 'Espresso',         'price': 2.50, 'quantity': 1},
    {'name': 'Blueberry Muffin', 'price': 2.00, 'quantity': 1},
]
```

### Step 2 — Subtotal

Loop reads `price * quantity` for each line and adds it to a running total.

```python
subtotal = 0
for item in order:
    subtotal += item['price'] * item['quantity']

print(f"Subtotal: €{subtotal:.2f}")
# 2*3.75 + 1*2.50 + 1*2.00 = €12.00
```

### Step 3 — Discount

€12.00 is over €10, so 10% kicks in → €1.20 off → €10.80 total.

```python
discount = 0
if subtotal > 10:
    discount = subtotal * 0.10

total = subtotal - discount
print(f"Discount: €{discount:.2f}")
print(f"Total:    €{total:.2f}")
```

### Step 4 — Free muffin check

3 coffees (2 lattes + 1 espresso) → just hits the threshold.

```python
coffee_names = ['Latte', 'Cappuccino', 'Espresso']
coffee_count = 0
for item in order:
    if item['name'] in coffee_names:
        coffee_count += item['quantity']

if coffee_count >= 3:
    print(f"🎉 {coffee_count} coffees ordered — free muffin!")
else:
    print(f"{coffee_count} coffees ordered. Need {3 - coffee_count} more for a free muffin.")
```

### Bonus — Reusable function

All logic in one place. Now `checkout(any_order)` works.

```python
def checkout(order):
    coffee_names = ['Latte', 'Cappuccino', 'Espresso']
    subtotal = sum(item['price'] * item['quantity'] for item in order)
    discount = subtotal * 0.10 if subtotal > 10 else 0
    coffee_count = sum(
        item['quantity'] for item in order if item['name'] in coffee_names
    )
    return {
        'subtotal':    round(subtotal, 2),
        'discount':    round(discount, 2),
        'total':       round(subtotal - discount, 2),
        'free_muffin': coffee_count >= 3,
    }

print(checkout(order))
print(checkout([{'name': 'Espresso', 'price': 2.50, 'quantity': 1}]))
```

### Teaching notes

- `sum(... for item in order)` is a **generator expression** — same as the for-loop, compact form. Show both.
- `round(x, 2)` matters for money — floating point can give you €10.799999...
- If a student asks: yes, `Decimal` is the proper money type. For teaching, `round()` is fine.
